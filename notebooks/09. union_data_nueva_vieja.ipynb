## Category

Creando las tablas de categoria para la data nueva

```python
category_explode = business_nuevo.assign(category=business_nuevo['category'].str.split(', ')).explode('category')
```

```python
business_category_nueva = category_explode[["category", "place_id", "google_id"]]
```

```python
business_category_nueva.rename(columns={"google_id":"business_id", "category":"category_name"}, inplace=True)
```

```python
business_category_nueva
```

```python
category_nuevo = pd.DataFrame()
category_nuevo["category_id"] = None
category_nuevo["category_name"] = business_category_nueva["category_name"].drop_duplicates()
```

```python
category_list = [category_viejo, category_nuevo]
category = pd.concat(category_list, ignore_index=True)
```

```python
category.drop_duplicates(subset="category_name", inplace=True)
```

```python
category.drop(columns="category_id", inplace=True)
```

```python
category = category.reset_index(drop=True)
category = category.reset_index()
category.rename(columns={"index":"category_id"}, inplace=True)
```

```python
category
```

# Tabla Calendario

```python
start_date = '2005-01-01'
end_date = '2023-12-31'
date_range = pd.date_range(start=start_date, end=end_date, freq='D')
```

```python
df = pd.DataFrame({'date': date_range})
df['year'] = df['date'].dt.year
df['quarter'] = df['date'].dt.quarter
df['month'] = df['date'].dt.month
```
