# Ultimando Detalles

Antes de exportar los dataframes a csv, se realizan unos cambios a la tabla de reviews. El primero, mapear los business_id borrados de Yelp porque estaban duplicados y reemplazarlos por los business_id nuevos.

```python
aux_table
```

```python
reviews.head()
```

```python
# Crear un diccionario de mapeo de business_id_y a business_id_x
mapping_dict = dict(zip(aux_table['business_id_y'], aux_table['business_id_x']))
```

```python
# Usar el método .map() para reemplazar los valores en reviews, y mantener los no mapeados
reviews['business_id'] = reviews['business_id'].apply(lambda x: mapping_dict.get(x, x))
```

Chequeamos que funciona

```python
reviews[reviews["business_id"] == "a9mX_pOERGedTzStIVhZCg"]
```

```python
reviews[reviews["business_id"] == "0x88d903c55457bd63:0xffd8bd14e832fb44"]
```

Se hace lo mismo con business_category

```python
business_category
```

```python
business_category['business_id'] = business_category['business_id'].apply(lambda x: mapping_dict.get(x, x))
```

Ahora se filtra la tabla reviews para que solo tenga datos de los estados de estudios, haciendo inner join con la tabla de hoteles.

```python
reviews = reviews.merge(business_filtrado["business_id"], on="business_id")
```

Como ya se concatenó la tabla transaccional de business_category, ahora ya se puede borrar la columna de business_id al igual que se hizo con la tabla de usuarios.

```python
business_filtrado.drop(columns="business_id_2", inplace=True)
```
