# Resto de estados

```python
def etl(df):
  drop_columns = ["query", "location_link",  "google_id", "reviews_link", "reviews", "rating", "reviews_per_score", "author_link",
                "author_image", "review_img_url", "review_img_urls", "owner_answer", "owner_answer_timestamp",
                "owner_answer_timestamp_datetime_utc", "review_link", "review_timestamp", "review_likes", "reviews_id"]

  df.drop(columns=drop_columns, inplace=True)
  df['review_datetime_utc'] = df['review_datetime_utc'].dt.date
  df.rename(columns={"name":"hotel_name",
                   "author_title": "user_name", "reviews": "num_of_reviews",
                   "author_id": "user_id", "review_text": "review", "review_rating": "rating",
                   "review_datetime_utc": "date"}, inplace=True)

  return df
```
