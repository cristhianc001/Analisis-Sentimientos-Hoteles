```python
def etl_hoteles(df):
  chosen_columns = ["name", "subtypes", "full_address", "state", "latitude", "longitude", "rating", "reviews", "place_id", "google_id"]
  df = df[chosen_columns]
  df.rename(columns={"name":"business_name", "subtypes": "category", "full_address": "address",
                    "rating": "avg_rating", "reviews": "num_of_reviews"}, inplace=True)
  return df
```

## Resto de estados

```python
def etl(df):
  drop_columns = ["query", "location_link", "reviews_link", "reviews", "rating", "reviews_per_score", "author_link",
                "author_image", "review_img_url", "review_img_urls", "owner_answer", "owner_answer_timestamp",
                "owner_answer_timestamp_datetime_utc", "review_link", "review_timestamp", "review_likes", "reviews_id"]

  df.drop(columns=drop_columns, inplace=True)
  df['review_datetime_utc'] = df['review_datetime_utc'].dt.date
  df.rename(columns={"name":"business_name",
                   "author_title": "user_name", "reviews": "num_of_reviews",
                   "author_id": "user_id", "review_text": "review", "review_rating": "rating",
                   "review_datetime_utc": "date"}, inplace=True)

  return df
```

```python
fl = etl(fl_df)
fl["state"] = "FL"
fl.head(3)
```

```python
nv = etl(nv_df)
nv["state"] = "NV"
nv.head(3)
```
