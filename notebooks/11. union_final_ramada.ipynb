Se hace un merge con hotel para agrupar por el nombre



```python
reviews_merged = reviews.merge(hotels_2, on="place_id")
reviews_merged.head()
```

```python
reviews_merged.isna().sum()
```

Se agrupa para calcular el rating promedio por hotel

```python
reviews_grouped = pd.DataFrame(reviews_merged.groupby(["hotel_name", "place_id", "state"])["rating"].mean().reset_index())
reviews_grouped.isna().sum()
```

```python
reviews_grouped
```

```python
reviews_grouped["rating"].fillna(3, inplace=True)
```

Ahora se hace otro merge con la tabla agrupada para reemplazar los rating faltantes por el promedio calculado

```python
reviews_2 = reviews_merged.merge(reviews_grouped, on="place_id")
```

```python
reviews_2.head(2)
```

```python
reviews_2.isna().sum()
```

```python
for index, row in reviews_2.iterrows():
    if pd.isnull(row['rating_x']):
        reviews_2.at[index, 'rating_x'] = row['rating_y']
```

```python
reviews_2.isna().sum()
```

Ahora se borran las columnas innecesarias

```python
chosen_columns = ["date", "rating_x","review", "user_id", "place_id"]
reviews_2 = reviews_2[chosen_columns]
reviews_2.rename(columns={"rating_x": "rating"}, inplace=True)
```

# Merged

Para producción es posible que se necesite una version denormalizada de estos datos

```python
hotels = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/union-total/hotels.csv", sep="|")
reviews = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/union-total/reviews.csv", sep="|", parse_dates=["date"])
users = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/union-total/users.csv", sep="|")
category = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/union-total/category.csv", sep="|")
hotels_category = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/union-total/hotels_category.csv", sep="|")
```

```python
category.isna().sum()
```

```python
category = category.fillna("Sin Categoria")
```

```python
place_id_category_name = hotels.merge(hotels_category, on="place_id", how="left").merge(category, on = "category_id", how="left") \
                          .fillna("Sin Categoria").groupby("place_id")["category_name"].agg(list).reset_index()
place_id_category_name
```

```python
place_id_category_name.isna().sum()
```

```python
hotels_denormalized = hotels.merge(place_id_category_name, on="place_id", how="left")
hotels_denormalized
```

```python
reviews_merge = reviews.merge(hotels_denormalized, on="place_id", how="left").merge(users, on="user_id", how="left")
```

```python
reviews_merge.isna().sum()
```

```python
reviews_merge.shape
```
